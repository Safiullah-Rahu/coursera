# Initial value problem for ODEs

# I. Euler's method for a first order ODE

Consider a first order equation

$$
\frac{d u}{d t} = \lambda u
$$

with the initial condition $u(t=0) = u_0$.

Here is a simple illustration of solving this equation with the explicit Euler method.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def euler_solve(lam, u0, T, dt):
    """Solve $du/dt = \lambda u$ on $0 < t < T$ with $u(t=0) = u0$ via an explicit Euler method."""
    num_steps = int(T/dt)
    tt = np.arange(num_steps+1)*dt
    y = np.empty(num_steps+1)
    y[0] = u0
    for k in range(num_steps):
        y[k+1] = y[k] + dt*lam*y[k]
    return tt, y
```

```python
lam = -0.5
tt, y = euler_solve(lam, u0=1.0, T=5, dt=0.1)
plt.figure(figsize = (14,10))
plt.plot(tt, y, 'o--', label='numeric solution')
plt.plot(tt, np.exp(lam*tt), '-', lw=2, label='ground truth')
plt.legend(loc='best')
plt.grid(True)
```

### Test I.1

Test the function above for varying step size $\tau$ (in the code it's `dt`), including $|\lambda| \tau > 1$? 

(10% of the grade)

```python
tau = [2**(-k+0.5) for k in range(-2, 5)]

plt.figure(figsize = (14,10))

for t in tau:
    tt, y = euler_solve(lam, u0=1.0, T=5, dt=t)
    plt.plot(tt, y, '.--', label='numeric solution, $\\tau = {tau}$'.format(tau = t), color='gray')
tt_exact = np.linspace(0, 5, 200)
plt.plot(tt_exact, np.exp(lam*tt_exact), 'r-', lw=2, label='ground truth')
```

### Test I.2

Implement a function for solving the same equation, $du/dt = \lambda u$ using the implicit Euler scheme. Compare the behavior of the implicit and explicit Euler schemes. Discuss.

(10% of the grade)

```python
def euler_implicit(lam, u0, T, dt):
    """Solve $du/dt = \lambda u$ on $0 < t < T$ with $u(t=0) = u0$ via an implicit Euler method."""
    num_steps = int(T/dt)
    tt = np.arange(num_steps+1)*dt
    y = np.empty(num_steps+1)
    y[0] = u0
    for k in range(num_steps):
        y[k+1] = y[k]/ (1-lam*dt)
    return tt, y

tau = [2**(-k+0.5) for k in range(-2, 5)]

def optional_label(tt, y, color, label, in_legend):
    if in_legend:
        plt.plot(tt, y, '.--', label=label, color=color)
    else:
        plt.plot(tt, y, '.--', color=color)
        
plt.figure(figsize = (14,10))       

for i, t in enumerate(tau):
    tt, y = euler_implicit(lam, u0=1.0, T=15, dt=t)
    optional_label(tt, y, color='gray',   label='implicit Euler', in_legend=(i==0))
    tt, y = euler_solve(lam, u0=1.0, T=15, dt=t)
    optional_label(tt, y, color='silver', label='explicit Euler', in_legend=(i==0))


tt_exact = np.linspace(0, 15, 200)
plt.plot(tt_exact, np.exp(lam*tt_exact), 'r-', lw=2, label='ground truth')
plt.legend()
```

## II. Stiff systems.

Consider a system of two first order equations

$$
\frac{d \mathbf{u} }{d t} = A \mathbf{u}
$$

where $\mathbf{u}$ is a two-dimensional vector, and $A$ is a known constant 2$\times$2 matrix.

Implement a generalization of `euler_solve` routine for solving a system of linear first-order equations with time-independent matrix $A$ using the explicit Euler's method.

```python
def euler_multidimension(A, u0, T, dt):
    """Solve $du/dt = \lambda u$ on $0 < t < T$ with $u(t=0) = u0$ via an explicit Euler method."""
    num_steps = int(T/dt)
    tt = np.arange(num_steps+1)*dt
    y = np.empty((num_steps+1, u0.size))
    y[0,:] = u0.copy()
    for k in range(num_steps):
        y[k+1,:] = y[k,:] + dt*A @ y[k,:]
    return tt, y
```

### Test II.1

Take 
$$
A = \begin{bmatrix} -10 & 10 \\ 32 & -499 \end{bmatrix}
$$

and the initial condition $\mathbf{u} = (1, 0)^T$.

Solve the system using a fixed step size $\tau=0.01$. Is the explicit Euler's method stable at this value of the step size?

Find eigenvalues of $A$ (use `np.linalg.eigvals`) and comment whether the system is stif.

(20% of the grade)

```python
u0 = np.array([1., 0.])
A = np.array([[-10,    10],
              [ 30,  -499]])
tau = 0.01
tt, y = euler_multidimension(A, u0, T=2, dt=tau)

fig, axs = plt.subplots(3)
axs[0].plot(y[:,0], y[:,1], '.--')
axs[1].plot(tt, y[:,0], '.--')
axs[2].plot(tt, y[:,1], '.--')
```

```python
np.linalg.eigvals(A)
```

### Since $499.61 \times 0.01 > 1$, Euler's explicit scheme is not stable for this step size. Moreover, the eigenvalue ratio is $\approx 50 \gg 1$, therefore the system can be called stiff.

### Test II.2

Implement the $\textit{implicit}$ Euler's scheme for a system of first-order equations with constant coefficients. Note that at each time step you need to solve a system of linear algebraic equations, use `np.linalg.solve` for that.

Use this routine to solve the system from Test II.1 at the same step size $\tau=0.01$. Compare solutions obtained by an explicit and an implicit Euler's methods.

(20% of the grade)

```python
def euler_multidim_implicit(A, u0, T, dt):
    """Solve $du/dt = \lambda u$ on $0 < t < T$ with $u(t=0) = u0$ via an explicit Euler method."""
    num_steps = int(T/dt)
    n = u0.size
    tt = np.arange(num_steps+1)*dt
    y = np.empty((num_steps+1, n))
    y[0,:] = u0.copy()
    for k in range(num_steps):
        y[k+1,:] = np.linalg.solve(np.eye(n) - dt * A, y[k,:])
    return tt, y

u0 = np.array([1., 0.])
A = np.array([[-10,    10],
              [ 30,  -499]])
t = 0.01
tt, y = euler_multidim_implicit(A, u0, T=2, dt=t)

fig, axs = plt.subplots(3)
axs[0].plot(y[:,0], y[:,1], '.--')
axs[1].plot(tt, y[:,0], '.--')
axs[2].plot(tt, y[:,1], '.--')
```

# III. Second order ODEs.

Consider a second order ODE, which describes a oscillating pendulum

$$
\frac{d^2 u}{dt^2} + \omega^2 u = 0
$$

Convert this second order ODE into a system of two first order ODEs.

### Test III.1 

Solve this system of equations using the explicit Euler's method over a time interval which includes at least several periods. We know that the equation of motion conserves energy, so that

$$
E = \frac{u'^2}{2} + \frac{\omega^2 u^2}{2}
$$

should remain constant. Plot the dependence of $E$ on time for your numeric solution. Use several values of the time step. Does your discretized scheme conserve energy?

(20% of the grade)

### <b>Answer:</b> That ODE can be rewritten as (slightly abusing notation)
$$ \frac{d}{dt} \begin{bmatrix} u \\ \dot u \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ - \omega^2 & 0 \end{bmatrix} \begin{bmatrix} u \\ \dot u \end{bmatrix} $$

Moreover, the period of this harmonic oscillator is $T = 2 \pi / \omega$, and the phase portrait should be an ellipse.

```python
omega = 1.3
A = np.matrix([[0, 1],[-omega**2, 0]])

def energy(q, p, omega):
    return .5 * p**2 + .5 * omega**2 * q**2

t = 0.01
tt, y = euler_multidimension(A, u0, T=32*np.pi/omega, dt=t)

fig, axs = plt.subplots(2, 2)
axs[0,0].plot(y[:,0], y[:,1], '-')
axs[1,0].plot(tt, y[:,0], '-')
axs[1,1].plot(tt, y[:,1], '-')
axs[0,1].plot(tt, energy(y[:,0], y[:,1], omega), 'r--')
axs[0,1].set_yscale('log')
```

### Implicit Euler scheme for comparison, even though this wasn't required

```python
omega = 1.3
A = np.matrix([[0, 1],[-omega**2, 0]])

def energy(q, p, omega):
    return .5 * p**2 + .5 * omega**2 * q**2

t = 0.01
tt, y = euler_multidim_implicit(A, u0, T=32*np.pi/omega, dt=t)

fig, axs = plt.subplots(2, 2)
axs[0,0].plot(y[:,0], y[:,1], '-')
axs[1,0].plot(tt, y[:,0], '-')
axs[1,1].plot(tt, y[:,1], '-')
axs[0,1].plot(tt, energy(y[:,0], y[:,1], omega), 'r--')
axs[0,1].set_yscale('log')
```

### Test III.2

Implement the 2nd order Runge-Kutta scheme. Use it to solve the same equation with same time steps. Compare solutions produced by the RK method and the Euler's method at the same values of the time step. Check conservation of energy. Discuss.

(20% of the grade)

### Solve $du/dt = \lambda u$ on $0 < t < T$ with $u(t=0) = u0$ via an explicit Euler method.

```python
def runge_kutta(A, u0, T, dt):
    num_steps = int(T/dt)
    n = u0.size
    tt = np.arange(num_steps+1)*dt
    y = np.empty((num_steps+1, n))
    y[0,:] = u0.copy()
    for k in range(num_steps):
        predictor = np.squeeze(np.array( y[k,:] + .5*dt* np.matmul(A, y[k,:]) ))
        y[k+1,:] = y[k,:] + dt*A @ predictor  # corrector
    return tt, y

omega = 1.3
A = np.matrix([[0, 1],[-omega**2, 0]])

def energy(q, p, omega):
    return .5 * p**2 + .5 * omega**2 * q**2

t = 0.1
tt, y = runge_kutta(A, u0, T=50*np.pi/omega, dt=t)

fig, axs = plt.subplots(2, 2)
axs[0,0].plot(y[:,0], y[:,1], '-')
axs[1,0].plot(tt, y[:,0], '-')
axs[1,1].plot(tt, y[:,1], '-')
axs[0,1].plot(tt, energy(y[:,0], y[:,1], omega), 'r--')
axs[0,1].set_ylim(0, 2*energy(y[0,0], y[0,1], omega))
```
